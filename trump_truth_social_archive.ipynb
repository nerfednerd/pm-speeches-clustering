```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('truth_archive.csv')
df
```

```python
# df.drop(columns=["id","url","media"], inplace=True)
display(df)
```

```python
df["created_at"]=pd.to_datetime(df["created_at"])
filtered_date_df = df[df['created_at'] > '2025-01-20']
filtered_date_df
```

```python
keywords=["Apple", "Tim Cook"]

pattern="|".join(keywords)
filtered_df_keyword=filtered_date_df[filtered_date_df['content'].str.contains(pattern, case=False,na=False)]
filtered_df_keyword= filtered_df_keyword.reset_index(drop=True)
display(filtered_df_keyword)
```

 sentiment analysis, only considering the created_at and content columns

```python
sentiment_df=filtered_df_keyword[["created_at","content"]]
display(sentiment_df)
```

```python
from transformers import pipeline

sentiment_pipe=pipeline("sentiment-analysis")

def get_sentiment(text):
  if not text or pd.isna(text):
    return {"label":None, "score":None}
  result=sentiment_pipe(text[:512])[0]
  return result


sentiment_df_v0=sentiment_df['content'].apply(get_sentiment)
sentiment_expanded = sentiment_df_v0.apply(pd.Series)

final_df = pd.concat([sentiment_df[['created_at', 'content']], sentiment_expanded], axis=1)

final_df = final_df.reset_index(drop=True)

final_df
```

```python
# Assume final_df has columns 'sentiment_label' and 'sentiment_score'
final_df['signed_sentiment_score'] = final_df.apply(
    lambda row: -row['score'] if row['label'] == 'NEGATIVE' else row['score'],
    axis=1
)
final_df

```

```python
import matplotlib.pyplot as plt

# Make sure 'created_at' is datetime
final_df['created_at'] = pd.to_datetime(final_df['created_at'])

# Sort by date (optional, for a proper time series)
final_df = final_df.sort_values('created_at')

# Plot
plt.figure(figsize=(12, 6))
plt.scatter(final_df['created_at'], final_df['signed_sentiment_score'], alpha=0.6)
plt.ylim(-1, 1)  # y-axis between -1 and 1
plt.xlabel('Created At')
plt.ylabel('Signed Sentiment Score')
plt.title('Sentiment over Time')
plt.xticks(rotation=45)  # rotate dates for readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

Aspect Based Sentiment Analysis

```python
from transformers import pipeline

# Aspect-Based Sentiment Analysis
absa_pipe = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-absa")

text = "Soybean Farmers ... Sleepy Joe Biden didn’t enforce our Agreement ..."
aspects = ["Biden", "farmers", "China"]

results = {aspect: absa_pipe(text + " [ASPECT] " + aspect) for aspect in aspects}
print(results)
```
